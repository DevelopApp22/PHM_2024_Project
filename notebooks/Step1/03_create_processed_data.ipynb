```python
from src.dataset import load_dataset,save_processed_dataset,create_features,create_np_ng_ratio
import pandas as pd
```

Creazione full_tarin


```python
TRAIN_PATH_X="../../data/raw/train/X_train.csv"
TRAIN_PATH_Y="../../data/raw/train/y_train.csv"

df_X_train=load_dataset(TRAIN_PATH_X)
df_y_train=load_dataset(TRAIN_PATH_Y)

df= pd.merge(df_X_train,df_y_train,on="id").drop("id", axis=1)
df=create_features(df)
save_processed_dataset(df, path="../../data/processed/full_train/train.csv")
```

Creazione dataset splittato

```python
import os
from sklearn.model_selection import train_test_split

target_col = "faulty"
output_folder = "../../data/processed/split_85_15"

os.makedirs(output_folder, exist_ok=True)
X = df.drop(columns=[target_col])
y = df[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.15,
    random_state=42,
    stratify=y
)
train_df = X_train.copy()
train_df[target_col] = y_train

test_df = X_test.copy()
test_df[target_col] = y_test

train_path = os.path.join(output_folder, "train_85.csv")
test_path = os.path.join(output_folder, "test_15.csv")

train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)
```
