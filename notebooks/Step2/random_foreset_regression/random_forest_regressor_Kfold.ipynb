```python
import pandas as pd
from regression.random_forest_regression import RandomForestModel
```

```python
TARGET_COL = "trq_margin"
TRAIN_PATH_X = "../../../data/processed/train/regression/train_trqMargin.csv"
```

```python
train_data = pd.read_csv(TRAIN_PATH_X)
drop_cols = [TARGET_COL]
X_cols = [c for c in train_data.columns if c not in drop_cols]
X_train= train_data[X_cols]
y_train = train_data[TARGET_COL]
```

```python
rf_model = RandomForestModel(
    n_estimators=300,
    min_samples_leaf=5
)
folds, summary = rf_model.cross_validate(X_train, y_train, n_splits=5)
```

```python
models_to_compare = [
    ("RF_leaf5", RandomForestModel(n_estimators=300, min_samples_leaf=5)),
    ("RF_leaf1", RandomForestModel(n_estimators=300, min_samples_leaf=1)),
    ("RF_depth10_leaf5", RandomForestModel(n_estimators=300, max_depth=10, min_samples_leaf=5)),
]

results = RandomForestModel.compare_cv(models_to_compare, X_train, y_train, n_splits=5)

for r in results:
    print(r["model"], "| RMSE:", r["RMSE_mean"], "±", r["RMSE_std"], "| R2:", r["R2_mean"])
```
